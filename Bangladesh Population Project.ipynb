```python

dist={'Division' :[ 'Dhaka','Khulna','Chattogram','Barishal','Mymensingh','Rajshahi','Rangpur','Sylhet'],'Population' : [44215107,17416645,33202326,9100102,12225498,20353119,17610956,11034863]}

df1 = pd.DataFrame.from_dict(dist)

df1["Population"]=df1["Population"]/1000000

df1.head(8)
```

```python
colors_list = ['firebrick','darkgoldenrod', 'gold', 'forestgreen', 'slateblue', 'coral','royalblue','yellowgreen']
# explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge.
labels1=[ 'Dhaka','Khulna','Chattogram','Barishal','Mymensingh','Rajshahi','Rangpur','Sylhet']

df1['Population'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=labels1,         # turn off labels on pie chart
                            pctdistance=0.75,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list,  # add custom colors
                             
                            )

# scale the title up by 12% to match pctdistance
plt.title('According to the population division of Bangladesh', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df1.index, loc='upper left') 

plt.show()
```
